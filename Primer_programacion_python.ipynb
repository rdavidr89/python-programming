```python
#Inicio del programa
#El documento se tiene que descargar de la direccion https://www.kaggle.com/lalish99/covid19-mx
#El arhivo de dicha direccion es casos_confirmados.csv
#Por favor cargar el archivo en algunas de las carpetas de "Archivos" que se observa en el lado izquiedo
#Para cargar el arhivo poner el puntero del mouse en la carpeta deseada y se observara 3 puntos
#Hacer click en los puntos y click en subir. Cargar archivo
#Ya el arhivo cargado, abrir carpeta dandole click
#El archivo se observara. Apuntar al archivo "casos_confirmados.csv". Nuevamente, darle click en los 3 puntos y click en copiar ruta
#Correr el codigo y copiar enlace en "dame una direccion para funcionar". Dar enter
print('Bienvenido al codigo casos_confirmados.csv')
file_handler=open(input('dame una direccion para funcionar:'))
contar=0
masculino=0
femenino=0
promedio=0
elementos_lista=0
for lista in file_handler:
    datos=lista.split(',')
    if contar > 0:
        print(datos[2:4])
        promedio=promedio+int(datos[3])
        elementos_lista=elementos_lista+int(len(lista[3]))
    else:
        contar=contar+1
    if datos[2] =='MASCULINO':
        masculino=datos.count('MASCULINO')+masculino
    elif datos[2] == 'FEMENINO':
        femenino=datos.count('FEMENINO')+femenino

print('Numero de elementos:',elementos_lista)
promedio=promedio/elementos_lista
print('Promedio es =',int(promedio))
if femenino > masculino:
  print('La cantidad de Femenino con valor de', femenino, 'es mayor que la cantidad de Masculino con valor de',masculino)
else:
  print('La cantidad de Masculino con valor de', masculino, 'es mayor que la cantidad de Femenino con valor de',femenino)
```
